```python
# Importing of Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Loading the data
df = pd.read_csv('/Users/parek/OneDrive/Desktop/CODING/Advanced_Python/Libraries/MATPLOTLIB/graphs/gym_members_exercise_tracking_synthetic_data.csv')

# Clean column names: remove spaces and brackets
df.columns = df.columns.str.replace(" ", "_").str.replace("(", "").str.replace(")", "")

# Create new column: calories burned per hour
df["Calories_per_Hour"] = df["Calories_Burned"] / df["Session_Duration_hours"]

# -------------------- Analysis and Plots --------------------

# 1. Line chart: Average calories burned by workout type
avg_calories_by_workout = df.groupby("Workout_Type")["Calories_Burned"].mean()
plt.figure(figsize=(6, 4))
avg_calories_by_workout.plot(kind="line", marker='o', color="darkblue")
plt.title("Average Calories Burned by Workout Type (Line Chart)")
plt.xlabel("Workout Type")
plt.ylabel("Average Calories Burned")
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()

# 2. pie of calories burned by gender
calories_by_gender = df.groupby('Gender')['Calories_Burned'].sum()
plt.figure(figsize=(6, 6))
plt.pie(calories_by_gender, labels=calories_by_gender.index, autopct='%1.1f%%', startangle=90)
plt.title("Total Calories Burned by Gender")
plt.axis('equal')  # Keep pie chart circular
plt.show()

# 3. Histogram of calories burned per hour
plt.figure(figsize=(6, 4))
plt.hist(df["Calories_per_Hour"], bins=20, color="salmon", edgecolor="black")
plt.title("Distribution of Calories Burned per Hour")
plt.xlabel("Calories per Hour")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha = 0.5)
plt.tight_layout()
plt.show()

# 4. Bar Chart: Average calories burned by experience level 
avg_calories_by_experience = df.groupby("Experience_Level")["Calories_Burned"].mean()
plt.figure(figsize=(6, 4))
avg_calories_by_experience.plot(kind="bar", color="lightgreen")
plt.title("Avg Calories Burned by Experience Level")
plt.xlabel("Experience Level")
plt.ylabel("Calories Burned")
plt.grid(axis='y', linestyle='--', alpha = 0.5)
plt.tight_layout()
plt.show()

# 5. Calories burned vs. Session duration (scatter plot)
plt.figure(figsize=(6, 4))
plt.scatter(df["Session_Duration_hours"], df["Calories_Burned"], color="purple")
plt.title("Calories Burned vs. Session Duration")
plt.xlabel("Session Duration (hours)")
plt.ylabel("Calories Burned")
plt.grid(True, linestyle='--', alpha = 0.3)
plt.tight_layout()
plt.show()
```
